### Задача №1.

Для набора данных проведите кодирование одного (произвольного) категориального признака с использованием метода "count (frequency) encoding".

```python
enc_nom_1 = (train.groupby('POSTED_BY').size()) / len(train)
enc_nom_1
```

```python
train['POSTED_BY_ENCODE'] = train['POSTED_BY'].apply(lambda x : enc_nom_1[x])
train.head()
```

### Задача №21.

Для набора данных проведите масштабирование данных для одного (произвольного) числового признака с использованием масштабирования по медиане.

```python
from sklearn.preprocessing import RobustScaler

data_scaled_temp = RobustScaler().fit_transform(train[["TARGET(PRICE_IN_LACS)", "SQUARE_FT"]])

new_df = pd.DataFrame(data_scaled_temp)
new_df.columns=["TARGET_SCALED", "SQUARE_FT_SCALED"]

print(new_df)
# data_cs41_scaled = arr_to_df(data_cs41_scaled_temp)
# data_cs41_scaled.describe()
```

```python
new_train = train.join(new_df)
new_train = new_train[new_train["TARGET_SCALED"]<200]
```
